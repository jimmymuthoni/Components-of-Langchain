## Ollama
Ollama support embedding models, making it possible to build retrival augmented generation(RAG) applications that combine text prompts with existing documents or other data.

```python
from langchain_community.embeddings import OllamaEmbeddings
embeddings = (
    OllamaEmbeddings(model="gemma:2b") #by default it uses llama2
)
embeddings
```

```python
sent1 = embeddings.embed_documents(
    [
        "Alpha is the fisrt letter of Greek alphabet",
        "Beta is the secaond letter of Greek alphabet",
    ]
)
```

```python
# query
embeddings.embed_query("What is the second letter of Greek alphabet")
```
